```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
counts = titanic['pclass'].value_counts()
counts
```

```python
fig, ax = plt.subplots()
ax.pie(counts)
plt.show()
```

```python
print(counts.index)
```

```python
counts = counts.sort_index()
counts.index
```

```python
fig, ax = plt.subplots()
ax.pie(counts, labels = ['1st', '2nd', '3rd'])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(counts, autopct = '%1.1f%%')
ax.legend(labels = ['1st', '2nd', '3rd'], title = 'Titanic Class', loc = 'lower right')
ax.set_title("Titanic Passenger Class Breakdown")
plt.show()
```

```python
#Matplotlib Scatterplots
```

```python
titanic_notna = titanic[titanic['age'].notna()]
fig, ax = plt.subplots()
ax.scatter('age', 'fare', data = titanic_notna)
plt.show()
```

```python
female = titanic_notna[titanic_notna['sex'] == 'female']
male = titanic_notna[titanic_notna['sex'] == 'male']

fig, ax = plt.subplots()

scatter_f = ax.scatter('age', 'fare', marker = 'o', c = 'pclass', label = 'female', cmap = 'Paired', 
                       alpha = .8, data = female)

scatter_m = ax.scatter('age', 'fare', marker = '^', c = 'pclass', label = 'male', cmap = 'Paired', 
                       alpha = .8, data = male)

plt.title('Titanic Prices By Age Graph')

ax.set(xlabel = 'Age', ylabel = 'Fare')
ax.legend(*scatter_f.legend_elements())

plt.show()
```

```python
dots = sns.load_dataset("dots")
sns.relplot(x="time", y="firing_rate", col="align", hue="choice", size="coherence", style="choice",
           facet_kws=dict(sharex=False), kind="line", legend="full", data=dots)

plt.show()
```

```python
sns.set()
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", col="time", hue="smoker", style="smoker", size="size", data=tips);
```

```python
#Seaborn Distribution Plots (with Kernel Density Estimation KDE curve on top)
```

```python
sns.distplot(titanic['age'].dropna())
plt.show()
```

```python
sns.histplot(titanic['age'].dropna())
plt.show()
```

```python
sns.distplot(titanic['age'].dropna(), rug = True)
plt.show()
```

```python
from scipy.stats import norm

fig, ax = plt.subplots(figsize = (8, 5))

sns.distplot(titanic['age'].dropna(),
            fit = norm,
            kde = False,
            rug = True,
            hist = False,
            color = 'y')

plt.show()
```
